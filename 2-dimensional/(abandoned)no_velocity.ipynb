```python
def func_a(x):
    if x==0:
        return 0
    else:
        return x/(2*x+1)

def func_b(x):
    if x==0:
        return 0
    else:
        return (x+1)/(2*x+1)

def func_g(x):
    if x==0:
        return 0
    else:
        return x*(x+1)/(2*x+1)

def func_h(x):
    if x==0:
        return 0
    else:
        return x*(x+1)
    
def func_i(x):
    if x==0:
        return 0
    else:
        return 1/(2*x+1)

def fn_2(x):
    ans = func_b(x-2)*func_b(x-1) - 2*func_b(x-2)*func_i(x-1) - func_h(x-2)*func_i(x-2)*func_i(x-1)
    return ans

def fn(x):
    ans = 2*( func_a(x)*func_i(x-1) -func_b(x)*func_i(x+1) ) - 1*( func_a(x)*func_b(x-1) + func_a(x+1)*func_b(x) ) \
        - 1*func_h(x)*func_i(x)*( func_i(x-1)+func_i(x+1) )
    return ans

def fn2(x):
    ans = func_a(x+2)*func_a(x+1) + 2*func_a(x+2)*func_i(x+1) - func_h(x+2)*func_i(x+2)*func_i(x+1)
    return ans

def fn_all(x):
    ans = fn_2(x+2) + fn(x) + fn2(x-2)
    return ans
```
